```python
#Import UHQ data into a Pandas Dataframe
import tomotopy as tp
import little_mallet_wrapper
import seaborn
import glob
from pathlib import Path
import pandas as pd

#Load UHQ data into a pandas dataframe
df = pd.read_excel("../data/UHQ_FULL_EDITED.xlsx")
df
```

```python
from nltk.corpus import stopwords

year_begin = 1928 #UHQ first year is 1928
year_end = 2021 #UHQ end year is 2021

#creating string containing all UHQ texts
topic_model_df = df[(df['text'].notna()) & (df['year'] >= year_begin) & (df['year'] <= year_end)].reset_index()

texts = topic_model_df.text.tolist()
titles = topic_model_df.titles.tolist()

stop_words = ["one", "two", "also", "came", "well", "however", "often", "use", 
             "could", "went", "would", "many", "said", "told", "get", "going", "know",
             "might", "because", "made", "could", "yet", "much", "even", "many",
             "still", "made", "make", "good", "like", "must", "come", "got", "without",
             "called", "within", "three", "first", "began", "take"] + list(stopwords.words('english'))


training_data = []
for t in texts:
    processed_text = little_mallet_wrapper.process_string(t, numbers='remove',
                                                          lowercase=True, 
                                                          remove_short_words=True, 
                                                          remove_stop_words=True,
                                                          remove_punctuation=True, 
                                                          stop_words=stop_words
                                                         )
    training_data.append(processed_text)

little_mallet_wrapper.print_dataset_stats(training_data)
```

```python
# Number of topics to return
num_topics = 70

# Numer of topic words to print out
num_topic_words = 10

# Intialize the model
model = tp.LDAModel(k=num_topics)

# Add each document to the model, after splitting it up into words
for text in training_data:
    model.add_doc(text.strip().split())
    
print("Topic Model Training...\n\n")
# Iterate over the data 10 times
iterations = 10
for i in range(0, 100, iterations):
    model.train(iterations)
    print(f'Iteration: {i}\tLog-likelihood: {model.ll_per_word}')

print("\nTopic Model Results:\n\n")
# Print out top 10 words for each topic
topics = []
topic_individual_words = []
for topic_number in range(0, num_topics):
    topic_words = ' '.join(word for word, prob in model.get_topic_words(topic_id=topic_number, top_n=num_topic_words))
    topics.append(topic_words)
    topic_individual_words.append(topic_words.split())
    print(f"✨Topic {topic_number}✨\n\n{topic_words}\n")
```

```python
topic_distributions = [list(doc.get_topic_dist()) for doc in model.docs]
```

```python
from IPython.display import Markdown, display
import re

def make_md(string):
    display(Markdown(str(string)))

def get_top_docs(docs, topic_distributions, topic_index, n=5):
    
    sorted_data = sorted([(_distribution[topic_index], _document) 
                          for _distribution, _document 
                          in zip(topic_distributions, docs)], reverse=True)
    
    topic_words = topics[topic_index]
    
    make_md(f"### ✨Topic {topic_index}✨\n\n{topic_words}\n\n---")
    
    for probability, doc in sorted_data[:n]:
        # Make topic words bolded
        for word in topic_words.split():
            if word in doc.lower():
                doc = re.sub(f"\\b{word}\\b", f"**{word}**", doc, re.IGNORECASE)
        
        make_md(f'✨  \n**Topic Probability**: {probability}  \n**Document**: {doc}\n\n')
    
    return
```

```python
# Lists the top ten documents most associated with this topic. 

topic_num = 0
get_top_docs(titles, topic_distributions, topic_index=topic_num, n=10)
```

```python
#Here we calculate the topic probability score for each topic and each document.

import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)

topic_results = []
for topic_distribution in topic_distributions:
    topic_results.append({'topic_distribution': topic_distribution})

topics_df = pd.DataFrame(topic_results)
column_names = [f"Topic {number}" for number, topic in enumerate(topic_individual_words)]
topics_df[column_names] = pd.DataFrame(topics_df['topic_distribution'].tolist(), index = topics_df.index)
topics_df = topics_df.drop('topic_distribution', axis='columns')
topics_df['Dominant_topic'] = (np.argmax(topics_df.values, axis=1))

full_topics_model_df = pd.merge(topic_model_df, topics_df, left_index = True, right_index = True)
full_topics_model_df
```

```python
#We can view which topics are most dominant in our corpus. 
full_topics_model_df['Dominant_topic'].value_counts()
```

```python
#We can view all of the articles that had a given topic as its most dominant topic. 

topic_number = 0

full_topics_model_df[full_topics_model_df.Dominant_topic == topic_number]
```

```python
#Finally, we can look at the distribution of a given topic over time 
# In this case we are only plotting the number of articles where this was the dominant topic by year. 

topic_number = 16

topic_graph = full_topics_model_df[full_topics_model_df.Dominant_topic == topic_number]

output = topic_graph.groupby(['year']).count()
output['Dominant_topic'].plot()
```

```python
# Of course, often a given topic will still appear in an article even though it isn't the dominant topic. 
# Here we are plotting the sum of the percentage that a given topic appears across all articles by year. 

topic_number = 16

output = topic_graph.groupby(['year']).sum()
output["Topic "+ str(topic_number)].plot()
```
