## UHQ Networks

### Spatial Networks

```python
#Load UHQ data into a pandas dataframe
df = pd.read_excel("../UHQ_FULL_EDITED.xlsx")
```

```python
#Look at the first article in our database
text = df.text.tolist()[0]

#Run Named Entity Recognition on the text
document = nlp(text)

displacy.render(document[:500], style="ent")
```

```python
#Create a list of all location (GPE) entities in the text
location_list = []
for named_entity in document.ents:
    if named_entity.label_ == "GPE":
        location_list.append(named_entity.text.replace("’s", ""))

#Print our list
location_list
```

```python
#We can use the geolocator library to locate the latitude and longitude for each location. 
geolocator = Nominatim(user_agent="YOUR NAME's mapping app", timeout=2)

location_tally = Counter(location_list)

location_df = pd.DataFrame(location_tally.most_common(), columns=['location', 'count'])

loc_list = location_df.location.tolist()

lat = []
lon = []

for l in loc_list:
    temp_loc = geolocator.geocode(l)
    lat.append(temp_loc.latitude)
    lon.append(temp_loc.longitude)

location_df['lat'] = lat
location_df['lon'] = lon
location_df
```

```python
nodes_list = location_df.location.tolist()

edge1 = []
edge2 = []
weight = []

for e1 in range(len(nodes_list)):
    for e2 in range(len(nodes_list)-e1-1):
        edge1.append(nodes_list[e1])
        edge2.append(nodes_list[e2+e1+1])
        if nodes_list[e1] in text and nodes_list[e2+e1+1] in text:
            weight.append(1)

edges_df = pd.DataFrame()
edges_df['Source'] = edge1
edges_df['Target'] = edge2
edges_df['Weight'] = weight
edges_df
```

```python
G = nx.from_pandas_edgelist(edges_df, 'Source', 'Target', 'Weight')
```

```python
plt.figure(figsize=(8,8))
nx.draw(G, with_labels=True, node_color='skyblue', width=.3, font_size=8)
```

### STATES IN UHQ

```python
states_nodes = pd.read_csv("../spatial history/states.csv")
states_nodes
```

```python
texts = df[df.text.notna()].text.tolist()

snlist = states_nodes.name.tolist()
edge1 = []
edge2 = []
weight = []

for e1 in range(len(snlist)):
    for e2 in range(len(snlist)-e1-1):
        edge1.append(snlist[e1])
        edge2.append(snlist[e2+e1+1])
        sub_weight = 0
        for t in texts:
            if snlist[e1] in t and snlist[e2+e1+1] in t:
                sub_weight += 1      
        weight.append(sub_weight)

states_edges_df = pd.DataFrame({
            "Source": edge1,
            "Target": edge2,
            "Weight": weight
})

states_edges_df[states_edges_df.Weight > 0]
```

```python
sub_df = states_edges_df[states_edges_df.Weight > 0]
G = nx.from_pandas_edgelist(sub_df, 'Source', 'Target', 'Weight')
plt.figure(figsize=(8,8))
nx.draw(G, with_labels=True, node_color='skyblue', width=sub_df.Weight/100, font_size=8)
```

```python
plt.figure(figsize = (10,9))
m = Basemap(
         projection='merc',
         llcrnrlon=-180,
         llcrnrlat=10,
         urcrnrlon=-50,
         urcrnrlat=70,
         lat_ts=0,
         resolution='l',
         suppress_ticks=True)
```

```python
mx, my = m(states_nodes['longitude'].values, states_nodes['latitude'].values)
pos = {}
for count, elem in enumerate (states_nodes['name']):
     pos[elem] = (mx[count], my[count])
```

```python
plt.figure(figsize = (20,16))
nx.draw_networkx_nodes(G = G, pos = pos, nodelist = G.nodes(), 
                        node_color = 'r', alpha = 0.8, node_size = 100)
nx.draw_networkx_edges(G = G, pos = pos, edge_color='g', width=sub_df.Weight/50,
                        alpha=0.2, arrows = False)
m.drawcountries(linewidth = 1)
m.drawstates(linewidth = 0.2)
m.drawcoastlines(linewidth=3)
plt.tight_layout()
plt.show()
```

```python
people_edges_df = pd.read_csv("people_edges_50.csv")
sub_df = people_edges_df[people_edges_df.Weight > 0]
G = nx.from_pandas_edgelist(sub_df, 'Source', 'Target', 'Weight')
plt.figure(figsize=(8,8))
nx.draw(G, with_labels=True, node_color='skyblue', width=sub_df.Weight/50, font_size=8)
```

```python
#Degree -- i.e. how many connections does each node have? 
nx.degree(G)
```

```python
#Weighted Degree -- adds up total edge weights for a given node
nx.degree(G, weight='Weight')
```

```python
#Calculate Betweenness Centrality -- how often is a node a "bridge" connecting other nodes?
nx.betweenness_centrality(G)
```
