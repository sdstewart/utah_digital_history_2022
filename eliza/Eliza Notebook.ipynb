## Question 1:
Who is writing the UHQ? Breaking down into gender (and looking at the eras of editorship like Jed Rogers suggested), what is the gender parity of authorship at UHQ?

```python
#Import UHQ data into a Pandas Dataframe
import pandas as pd

#Load UHQ data into a pandas dataframe
df = pd.read_excel("../data/UHQ_FULL_EDITED.xlsx")
df
```

```python
#Create a list of all of the authors with predicted gender and year of publication.

df2 = df[df['gender'].notna()]
gender_list = df2['gender'].tolist()
author_list = df2['author'].tolist()
year_list = df2['year'].tolist()

full_gender_list = []
full_author_list = []
full_year_list = []

for g, a, y in zip(gender_list, author_list, year_list):
    for tg, ta in zip(g.split(';'), a.split(';')):
        full_gender_list.append(tg.strip())
        full_author_list.append(ta.strip())
        full_year_list.append(y)

gender_df = pd.DataFrame({
    'year': full_year_list,
    'gender': full_gender_list,
    'name': full_author_list
})

gender_df
```

```python
#Here we are creating a new dataframe. cross_tab_prop shows the percentage of man-woman authorship by year. 
cross_tab_prop = pd.crosstab(index=gender_df['year'],
                             columns=gender_df['gender'],
                             normalize="index")

cross_tab_prop
```

```python
#We can use this new cross_tab_prop dataframe to plot the male-female ratio by year over time. 
import matplotlib.pyplot as plt

color = 'inferno' #List of colors can be found here: https://matplotlib.org/stable/tutorials/colors/colormaps.html

cross_tab_prop.plot(kind='bar', 
                    stacked=True, 
                    colormap=color, 
                    figsize=(20, 14))

plt.legend(loc="upper right", ncol=3)
plt.xlabel("Release Year")
plt.ylabel("Proportion")
plt.show()
```

```python
#We can then single out individual years based on percentage of articles authored by women. 
percentage = 0.5

cross_tab_prop[cross_tab_prop['female'] > percentage]
```

```python
#We can view the individual articles found in specific years such as 1942 or 1970 or ...

year = 1942

df[df.year == year]
```

```python
## We can adjust the year_start and year_end variable to explore gender composition by different periods of time. 
## First we have the raw counts (i.e. an author such as Leonard Arrington will be counted multiple times here)

year_start = 1928
year_end = 2021

cross_tab = pd.crosstab(index=gender_df['year'], columns=gender_df['gender'])

sub_df = cross_tab[(cross_tab.index >= year_start) & (cross_tab.index <= year_end)]

print(f'Raw author counts from {year_start}-{year_end}: ')
print(f'Women author count: {sub_df.female.sum()}')
print(f'Men author count: {sub_df.male.sum()}')
print(f'Unknown author count: {sub_df["?"].sum()}')
print(f'Total author count: {(sub_df.female.sum() + sub_df.male.sum() + sub_df["?"].sum())}')

print(f'\nWomen percentage of total: {sub_df.female.sum() / (sub_df.female.sum() + sub_df.male.sum() + sub_df["?"].sum())}')
print(f'Men percentage of total: {sub_df.male.sum() / (sub_df.female.sum() + sub_df.male.sum() + sub_df["?"].sum())}')
print(f'Unknown percentage of total: {sub_df["?"].sum() / (sub_df.female.sum() + sub_df.male.sum() + sub_df["?"].sum())}')
```

```python
## Another approach is to only count each author once. 

authors = []
gender = []
count = []

sub_df = gender_df[(gender_df.year >= year_start) & (gender_df.year <= year_end)]

for a, g in zip(sub_df.name.tolist(), sub_df.gender.tolist()):
    if a not in authors:
        authors.append(a)
        gender.append(g)
        count.append(sub_df.name.tolist().count(a))
        
unique_authors = pd.DataFrame(
            {"author": authors,
            "gender": gender,
            "count": count}
            )

print(f'Unique authors from {year_start}-{year_end}: ')
print(f'Women author count: {len(unique_authors[unique_authors.gender == "female"])}')
print(f'Men author count: {len(unique_authors[unique_authors.gender == "male"])}')
print(f'Unknown author count: {len(unique_authors[unique_authors.gender == "?"])}')
print(f'Total author count: {len(unique_authors)}')

print(f'\nWomen percentage of total: {len(unique_authors[unique_authors.gender == "female"]) / len(unique_authors)}')
print(f'Men percentage of total: {len(unique_authors[unique_authors.gender == "male"]) / len(unique_authors)}')
print(f'Unknown percentage of total: {len(unique_authors[unique_authors.gender == "?"]) / len(unique_authors)}')

average_women = sum(unique_authors[unique_authors.gender == "female"]['count'].tolist()) / len(unique_authors[unique_authors.gender == "female"])
average_men = sum(unique_authors[unique_authors.gender == "male"]['count'].tolist()) / len(unique_authors[unique_authors.gender == "male"])
print(f'\nAverage # of times women authors are publishing: {average_women}')
print(f'Average # of times men authors are publishing: {average_men}')

print('\nThe top 10 women authors were:')
print(unique_authors[unique_authors.gender == "female"].sort_values(by=['count'], ascending=False)[:10])

print('\nThe top 10 men authors were:')
print(unique_authors[unique_authors.gender == "male"].sort_values(by=['count'], ascending=False)[:10])
```

### UHQ TOPIC MODEL

```python
import tomotopy as tp
import little_mallet_wrapper
import seaborn
import glob
from pathlib import Path
```

```python
year_begin = 1928 #UHQ first year is 1928
year_end = 2021 #UHQ end year is 2021

#creating string containing all UHQ texts
topic_model_df = df[(df['text'].notna()) & (df['year'] >= year_begin) & (df['year'] <= year_end)].reset_index()

texts = topic_model_df.text.tolist()
titles = topic_model_df.titles.tolist()

training_data = []
for t in texts:
    processed_text = little_mallet_wrapper.process_string(t, numbers='remove')
    training_data.append(processed_text)

little_mallet_wrapper.print_dataset_stats(training_data)
```

Before looking at specific topics, here I am using perplexity score to calculate the "ideal" number of topics. We can then check this out manually by looking at the topics to see if the score makes sense. Low perplexity score means a better fit.
```python
# Load out perplexity table and sort them from low to high. Note that we are looking for low perplexity scores. 
# k = number of topics

perplexity_df = pd.read_csv("../data/uhq_perplexity.csv", index_col=0)

perplexity_df.sort_values(by=['perplexity'])[:20]
```

```python
# Number of topics to return
num_topics = 74

# Numer of topic words to print out
num_topic_words = 10

# Intialize the model
model = tp.LDAModel(k=num_topics)

# Add each document to the model, after splitting it up into words
for text in training_data:
    model.add_doc(text.strip().split())
    
print("Topic Model Training...\n\n")
# Iterate over the data 10 times
iterations = 10
for i in range(0, 100, iterations):
    model.train(iterations)
    print(f'Iteration: {i}\tLog-likelihood: {model.ll_per_word}')

print("\nTopic Model Results:\n\n")
# Print out top 10 words for each topic
topics = []
topic_individual_words = []
for topic_number in range(0, num_topics):
    topic_words = ' '.join(word for word, prob in model.get_topic_words(topic_id=topic_number, top_n=num_topic_words))
    topics.append(topic_words)
    topic_individual_words.append(topic_words.split())
    print(f"✨Topic {topic_number}✨\n\n{topic_words}\n")
```

```python
topic_distributions = [list(doc.get_topic_dist()) for doc in model.docs]
```

```python
from IPython.display import Markdown, display
import re

def make_md(string):
    display(Markdown(str(string)))

def get_top_docs(docs, topic_distributions, topic_index, n=5):
    
    sorted_data = sorted([(_distribution[topic_index], _document) 
                          for _distribution, _document 
                          in zip(topic_distributions, docs)], reverse=True)
    
    topic_words = topics[topic_index]
    
    make_md(f"### ✨Topic {topic_index}✨\n\n{topic_words}\n\n---")
    
    for probability, doc in sorted_data[:n]:
        # Make topic words bolded
        for word in topic_words.split():
            if word in doc.lower():
                doc = re.sub(f"\\b{word}\\b", f"**{word}**", doc, re.IGNORECASE)
        
        make_md(f'✨  \n**Topic Probability**: {probability}  \n**Document**: {doc}\n\n')
    
    return
```

```python
# Lists the top ten documents most associated with this topic. 

topic_num = 25
get_top_docs(titles, topic_distributions, topic_index=topic_num, n=10)
```

```python
#Here we calculate the topic probability score for each topic and each document.

import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)

topic_results = []
for topic_distribution in topic_distributions:
    topic_results.append({'topic_distribution': topic_distribution})

topics_df = pd.DataFrame(topic_results)
column_names = [f"Topic {number}" for number, topic in enumerate(topic_individual_words)]
topics_df[column_names] = pd.DataFrame(topics_df['topic_distribution'].tolist(), index = topics_df.index)
topics_df = topics_df.drop('topic_distribution', axis='columns')
topics_df['Dominant_topic'] = (np.argmax(topics_df.values, axis=1))

full_topics_model_df = pd.merge(topic_model_df, topics_df, left_index = True, right_index = True)
full_topics_model_df
```
